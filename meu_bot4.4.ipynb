```python
import subprocess
import pyautogui
import time
import pygetwindow as gw
from datetime import datetime

pyautogui.PAUSE = 0.5

subprocess.Popen(['notepad.exe'])

janela = None
inicio = time.time()

while time.time() - inicio < 10:
    janelas = gw.getWindowsWithTitle('Sem título - Bloco de Notas')
    if janelas:
        janela = janelas[0]
        break
    time.sleep(0.5)

if not janela:
    raise Exception("Não encontrei a janela do Bloco de Notas.")

try:
    janela.activate()
    time.sleep(1)
except:
    pass

pyautogui.click(janela.left + 200, janela.top + 200)
time.sleep(1)

agora = datetime.now().strftime("%d/%m/%Y %H:%M:%S")

texto = f"""Automacao com PyAutoGUI
Executado em: {agora}

vamos com tudoooo :)
"""

pyautogui.write(texto, interval=0.05)

time.sleep(1)
pyautogui.screenshot("snapshot_notepad.png")

pyautogui.hotkey('ctrl', 's')
time.sleep(2)

caminho = r"C:\Users\Public\Documents\teste_pyautogui.txt"
pyautogui.write(caminho, interval=0.05)

time.sleep(1)
pyautogui.press('enter')

time.sleep(2)
pyautogui.hotkey('alt', 'f4')
pyautogui.hotkey('alt', 'f4')
```
