### **Exercício 3: Abrir um Pop-up (chamado de modal), preencher e fechar sem usar o mouse**

Você vai criar um bot em Python usando PyAutoGUI para interagir com um elemento que aparece na tela (pop-up) e depois desaparece quando fechado.

Cenário: no formulário principal do site, existe um link chamado "Abrir Pop-up". Ao ativar esse link, aparece um pop-up com três campos (Campo 1, Campo 2 e um campo de Notas). Existe também um botão "Fechar".

**O que o bot deve fazer?**

1) Abra http://localhost:8000.

2) Navegue pelo formulário usando Tab até chegar no link "Abrir Pop-up".

3) Ative o link usando o teclado (Enter) (não use click).

4) Aguarde o pop-up aparecer.

5) Preencha os dois campos com algum texto; e feche o pop-up sem usar o mouse, colocando o foco no botão Fechar e ativando-o (Enter).

6) Ao final, o pop-up deve sumir e a página deve voltar ao campo "Observações".

**Restrições**

a) Não use de forma alguma clique do mouse.

b) Use Tab e Shift+Tab para navegar entre elementos.

c) Use Enter para ativar o link e o botão.

d) Use pausas quando necessário para garantir que o pop-up carregou antes de digitar.

e) Inclua comentários no código.

```python
import pyautogui
import webbrowser
import time

# ---------------------------------------------------------
# Configurações Iniciais
# ---------------------------------------------------------

# Adiciona uma pausa de 1 segundo entre cada comando do PyAutoGUI
pyautogui.PAUSE = 1.0

print("Iniciando o bot... Solte o teclado e o mouse.")

# 1) Abre o formulário local no navegador padrão
webbrowser.open('http://127.0.0.1:5500/index.html')

# Aguarda 5 segundos para garantir que a página carregou completamente
time.sleep(5)

# ---------------------------------------------------------
# Navegação até o link do Pop-up
# ---------------------------------------------------------

# 2) Navega pelo formulário usando Tab até chegar no link "Abrir Pop-up"
# OBS: Altere o número do range() abaixo para a quantidade exata de Tabs
# necessários para o cursor sair do início da página e chegar no link.
# (Vou supor que são 5 pulos)
for _ in range(13):
    pyautogui.press('tab')

# 3) Ative o link usando o teclado (Enter)
pyautogui.press('enter')

# 4) Aguarde o pop-up aparecer e a animação terminar
time.sleep(2)

# ---------------------------------------------------------
# Interação dentro do Pop-up (Modal)
# ---------------------------------------------------------

# Geralmente, ao abrir um modal, precisamos de um Tab para focar no primeiro campo interno
pyautogui.press('tab')

# 5) Preenche os dois campos principais
# Preenche o Campo 1
pyautogui.write('Informacao do Campo 1')

# Pula para o Campo 2
pyautogui.press('tab')

# Preenche o Campo 2
pyautogui.write('Informacao do Campo 2')

# Pula para o campo de Notas (o exercício não obriga a preencher, então só passamos por ele)
pyautogui.press('tab')

# Pula do campo de Notas para o botão "Fechar"
pyautogui.press('tab')

# Ativa o botão Fechar usando o Enter
pyautogui.press('enter')

# Pausa para dar tempo do pop-up sumir da tela
time.sleep(1.5)

# ---------------------------------------------------------
# Retornando ao campo "Observações"
# ---------------------------------------------------------

# 6) Fazer a página voltar ao campo "Observações".
# Dependendo do HTML, quando um modal fecha, o foco volta para o link que o abriu.
# Se o campo "Observações" estiver ANTES do link na tela, usamos Shift+Tab para voltar.
# Se estiver DEPOIS, usaríamos um Tab normal. 
# Aqui vou usar o Shift+Tab simulando que estamos voltando um campo:
pyautogui.hotkey('shift', 'tab')

print("Automação do pop-up concluída!")
```
