```python
import time
from pathlib import Path
from datetime import datetime
import pyautogui as pag

# Segurança ligada
pag.FAILSAFE = True
pag.PAUSE = 0.5

print("Começando em 3 segundos. Solte o mouse e teclado!")
time.sleep(3)

# ---------------------------------------------------------
# PASSO 1: Abrir via Menu Iniciar
# ---------------------------------------------------------
pag.press("win")
time.sleep(1.5)

pag.write("notepad")
time.sleep(1.5)

pag.press("enter")
time.sleep(2.0)

# ---------------------------------------------------------
# PASSO 2: Escrever a mensagem
# ---------------------------------------------------------
agora = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
mensagem = "Estudo de caso RPA - Windows + PyAutoGUI\n"
pag.write(mensagem, interval=0.01)
time.sleep(1.0)

# ---------------------------------------------------------
# PASSO 3: Criar a pasta e chamar o atalho de Salvar
# ---------------------------------------------------------
pasta_destino = Path.home() / "Documents" / "gds" / "pyautogui"
pasta_destino.mkdir(parents=True, exist_ok=True)
stamp = datetime.now().strftime("%Y%m%d_%H%M%S")
arquivo_txt = pasta_destino / f"caso_rpa_notepad_{stamp}.txt"

# AQUI ESTAVA O ERRO: Pressionamos e soltamos as teclas manualmente
# Isso garante que o Windows 11 registre o atalho Ctrl + Shift + S
pag.keyDown("ctrl")
pag.keyDown("shift")
pag.press("s")
pag.keyUp("shift")
pag.keyUp("ctrl")

# Pausa longa e absoluta para garantir que a janela "Salvar Como" apareceu
time.sleep(3.0) 

# ---------------------------------------------------------
# PASSO 4: Digitar o caminho e confirmar
# ---------------------------------------------------------
pag.write(str(arquivo_txt))
time.sleep(1.0)

pag.press("enter")
time.sleep(2.0)

# ---------------------------------------------------------
# PASSO 5: Fechar
# ---------------------------------------------------------
pag.hotkey("alt", "f4")

print("Concluído.")
print("Arquivo TXT salvo em:", arquivo_txt)
```
