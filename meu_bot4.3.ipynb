```python
import time
import subprocess
from pathlib import Path
from datetime import datetime
import pyautogui as pag
import os
import pyperclip  # Instale com: pip install pyperclip

# Configurações
pag.FAILSAFE = True
pag.PAUSE = 0.5

def abrir_notepad_seguro():
    """Tenta abrir o Bloco de Notas de várias formas"""
    try:
        subprocess.Popen(["notepad.exe"])
        time.sleep(2)
        return True
    except:
        pag.hotkey("win", "r")
        time.sleep(1)
        pag.write("notepad")
        pag.press("enter")
        time.sleep(2)
        return True

def verificar_se_arquivo_existe(caminho_arquivo):
    """Verifica se o arquivo já existe no disco"""
    return Path(caminho_arquivo).exists()

print("Começando em 3 segundos...")
time.sleep(3)

# Abrir Bloco de Notas
print("Abrindo Bloco de Notas...")
abrir_notepad_seguro()

# Escrever mensagem
agora = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
mensagem = f"Estudo de caso RPA - Windows + PyAutoGUI\nData e hora: {agora}\n"
pag.write(mensagem, interval=0.05)
time.sleep(0.5)

# Criar pasta se não existir
desktop = Path.home() / "Documents"
pasta_destino = desktop / "gds" / "pyautogui"
pasta_destino.mkdir(parents=True, exist_ok=True)

stamp = datetime.now().strftime("%Y%m%d_%H%M%S")
arquivo_txt = pasta_destino / f"caso_rpa_notepad_{stamp}.txt"

print(f"Tentando salvar em: {arquivo_txt}")

# ANTES DE SALVAR, verifica se o arquivo já existe no disco
arquivo_ja_existe = verificar_se_arquivo_existe(arquivo_txt)

# Salvar como
pag.hotkey("ctrl", "shift", "s")
time.sleep(2)

# Digitar caminho
pag.write(str(arquivo_txt))
time.sleep(1)

# Confirmar com Enter
pag.press("enter")
time.sleep(2)

# =========================================================
# ESTRUTURA DE DECISÃO BASEADA NO QUE SABEMOS ANTES
# =========================================================
if arquivo_ja_existe:
    print("⚠️  Arquivo já existia antes de salvar!")
    print("🖱️  A janela de substituição DEVE ter aparecido.")
    print("Pressionando Enter para confirmar substituição...")
    
    # A janela de substituição está com foco agora
    # Enter confirma "Sim"
    pag.press("enter")
    time.sleep(1)
    
    # Agora o arquivo foi substituído, podemos fechar o Bloco de Notas
    print("Fechando Bloco de Notas...")
    pag.hotkey("alt", "f4")
    time.sleep(1)
    
else:
    print("✅ Arquivo não existia antes. Arquivo novo criado!")
    print("Fechando Bloco de Notas...")
    
    # Arquivo novo, podemos fechar direto
    pag.hotkey("alt", "f4")
    time.sleep(1)
    
    # Se perguntar "Deseja salvar?" (não deveria, mas seguro)
    pag.press("n")
    time.sleep(0.5)

print("\n" + "="*50)
print("PROCESSO CONCLUÍDO!")
print("="*50)
print(f"Arquivo: {arquivo_txt}")

# Verificar
if arquivo_txt.exists():
    print(f"✓ Arquivo criado com sucesso! Tamanho: {arquivo_txt.stat().st_size} bytes")
    if arquivo_ja_existe:
        print("ℹ️  Um arquivo existente foi SUBSTITUÍDO.")
else:
    print(f"✗ Arquivo NÃO foi encontrado!")
```
