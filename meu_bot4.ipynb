```python
import time
from pathlib import Path
from datetime import datetime
import pyautogui as pag

# Segurança: mover o mouse para o canto superior esquerdo aborta
pag.FAILSAFE = True
pag.PAUSE = 0.2

print("Começando em 3 segundos...")
time.sleep(3)

# Abrir Bloco de Notas (Win+R -> notepad)
pag.hotkey("win", "r")
time.sleep(0.8)
pag.write("notepad")
pag.press("enter")
time.sleep(1.2)

# Escrever mensagem
agora = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
mensagem = "Estudo de caso RPA - Windows + PyAutoGUI\n"
pag.write(mensagem, interval=0.01)
time.sleep(0.3)

# Salvar no Documents do usuário
desktop = Path.home() / "Documents\gds\pyautogui"
stamp = datetime.now().strftime("%Y%m%d_%H%M%S")
arquivo_txt = desktop / f"caso_rpa_notepad_{stamp}.txt"

pag.hotkey("ctrl", "shift", "s")
time.sleep(2.0)

# digita o caminho completo e confirma
pag.write(str(arquivo_txt))
time.sleep(0.4)
pag.press("enter")
time.sleep(2.0)

# Fechar Bloco de Notas
pag.hotkey("alt", "a")
pag.press("s")
time.sleep(0.3)

print("Concluído.")
print("Arquivo TXT:", arquivo_txt)
```

```python
import time
import pygetwindow as gw

titulos_possiveis = ['Bloco de Notas', 'Notepad']

print('Aguardando o Notepad abrir... (Ctrl+C para parar)')

while True:
    notepad_aberto = any(gw.getWindowsWithTitle(t) for t in titulos_possiveis)

    if notepad_aberto:
        print('Notepad está aberto. Saindo do bot.')
        break

    time.sleep(0.5)
```
