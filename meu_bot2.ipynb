import pyautogui
import webbrowser
import time

# ---------------------------------------------------------
# Configurações Iniciais e Abertura do Site
# ---------------------------------------------------------

# Adiciona uma pausa de 1 segundo entre cada comando do pyautogui por segurança
# Isso ajuda a evitar que o bot digite mais rápido do que o navegador consegue processar
pyautogui.PAUSE = 1.0

print("Iniciando o bot... Não mexa no teclado ou mouse.")

# 1) Abre o formulário local no navegador padrão
webbrowser.open("http://127.0.0.1:5500/index.html")
# Pausa maior para garantir que a página carregue completamente
# (Ajuste esse tempo se a sua página demorar mais para abrir)
time.sleep(5)
''''''
# ---------------------------------------------------------
# Preenchimento dos Campos Iniciais
# ---------------------------------------------------------

# 2) Pressiona Tab para focar no primeiro campo (Nome)
# OBS: Se a sua página precisar de mais Tabs para chegar no campo Nome, 
# basta aumentar o número dentro do loop ou repetir o comando.
pyautogui.press('tab')

# Digita o Nome
pyautogui.write('Usuario de Teste')

# Pressiona Tab para ir para o próximo campo (Email)
pyautogui.press('tab')

# Digita o Email
pyautogui.write('teste.automacao@exemplo.com')

# ---------------------------------------------------------
# Seleção e Cópia usando Atalhos de Teclado
# ---------------------------------------------------------

# Pausa rápida para garantir que o texto do email terminou de ser digitado
time.sleep(1)

# 3.i) Seleciona todo o texto dentro do campo atual (Email)
# Usa o atalho 'Ctrl + A' (Selecionar Tudo)
pyautogui.hotkey('ctrl', 'a')
time.sleep(0.5) # Pausa para o sistema operacional registrar a seleção visualmente

# 3.ii) Copia o texto selecionado
# Usa o atalho 'Ctrl + C' (Copiar)
pyautogui.hotkey('ctrl', 'c')
time.sleep(0.5) # Pausa para garantir que o texto foi para a área de transferência

# ---------------------------------------------------------
# Navegação e Colagem
# ---------------------------------------------------------

# 4) Navega usando Tab até chegar no campo Observações
# OBS: O número de Tabs aqui depende da estrutura HTML do seu formulário.
# Supondo que 'Observações' seja o campo imediatamente após 'Email':
for _ in range(8): # Ajuste o número 8 para o número correto de campos entre Email e Observações 
    pyautogui.press('tab') 

# Se houver outros campos (ex: Telefone, Endereço) antes de Observações, 
# descomente e ajuste o código abaixo:
# for _ in range(3): # Troque 3 pelo número de campos intermediários
#     pyautogui.press('tab')

# 5) Cole o conteúdo copiado dentro do campo Observações
# Usa o atalho 'Ctrl + V' (Colar)
pyautogui.hotkey('ctrl', 'v')

print("Automação concluída com sucesso!")
