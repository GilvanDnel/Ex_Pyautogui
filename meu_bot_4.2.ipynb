```python
import time
import subprocess
from pathlib import Path
from datetime import datetime
import pyautogui as pag
import os

# Configurações
pag.FAILSAFE = True
pag.PAUSE = 0.5

def abrir_notepad_seguro():
    """Tenta abrir o Bloco de Notas de várias formas"""
    
    # Método 1: subprocess (mais confiável)
    try:
        subprocess.Popen(["notepad.exe"])
        time.sleep(2)
        return True
    except:
        pass
    
    # Método 2: Win+R
    pag.hotkey("win", "r")
    time.sleep(1)
    pag.write("notepad")
    pag.press("enter")
    time.sleep(2)
    
    # Clica na janela do Bloco de Notas
    pag.click(x=500, y=300)
    time.sleep(1)
    
    return True

print("Começando em 3 segundos...")
time.sleep(3)

# Abrir Bloco de Notas
print("Abrindo Bloco de Notas...")
abrir_notepad_seguro()

# Fechar qualquer janela do Chrome que possa estar aberta (opcional)
# pag.hotkey('alt', 'f4')  # Cuidado: isso fecharia a janela ativa

# Escrever mensagem
agora = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
mensagem = f"Estudo de caso RPA - Windows + PyAutoGUI\nData e hora: {agora}\n"
pag.write(mensagem, interval=0.05)
time.sleep(0.5)

# Criar pasta se não existir
desktop = Path.home() / "Documents"
pasta_destino = desktop / "gds" / "pyautogui"
pasta_destino.mkdir(parents=True, exist_ok=True)

stamp = datetime.now().strftime("%Y%m%d_%H%M%S")
arquivo_txt = pasta_destino / f"caso_rpa_notepad_{stamp}.txt"

print(f"Tentando salvar em: {arquivo_txt}")

# Salvar como
pag.hotkey("ctrl", "shift", "s")
time.sleep(2)

# Digitar caminho
pag.write(str(arquivo_txt))
time.sleep(1)

# Confirmar
pag.press("enter")
time.sleep(2)

# Confirmar substituição se necessário
pag.press("enter")
time.sleep(1)

# Fechar
pag.hotkey("alt", "f4")
time.sleep(1)

print("Concluído!")
print(f"Arquivo TXT salvo em: {arquivo_txt}")

# Verificar
if arquivo_txt.exists():
    print(f"✓ Arquivo criado com sucesso!")
else:
    print("✗ Arquivo NÃO foi encontrado!")
```
