```python
# ### **Exercício 1: Preencher o formulário até o campo "Observações"**

# Você vai criar um bot em Python usando a biblioteca PyAutoGUI para preencher automaticamente parte do formulário do site local.

# Cenário: o site está rodando em http://localhost:8000 e contém um formulário com vários campos. Esses campos podem ser acessados em sequência usando a tecla Tab.

# **O que o bot deve fazer:**

# 1) Abrir a página http://localhost:8000 no navegador (pode ser usando webbrowser.open() ou deixando o navegador aberto e só focando nele).

# 2) Colocar o foco no primeiro campo do formulário (Nome) usando Tab.

# 3) Preencher os seguintes campos, avançando sempre com Tab: Nome, Email, CPF, Telefone, Endereço, Cidade, Estado, CEP, Categoria e Observações.

# 4) Note que os campos Estado e Categoria são "seletores". No caso do Estado, selecione o valor AM usando a seta para baixo do teclado. No caso da Categoria, selecione o valor "Aluno", também usando a seta para baixo do teclado. Nos dois casos, confirme e avance para o próximo campo com Tab.

# 5) Ao final, o cursor deve estar no campo "Observações".

# **Restrições**

# a) Não use o mouse (sem click() para focar campos).

# b) A navegação deve ser feita com Tab e as setas para o seletor.

# c) Use pausas (time.sleep() ou pyautogui.PAUSE) para evitar possíveis falhas.

# d) Escolhe os textos a serem inseridos, desde que pareçam válidos.


import pyautogui
import webbrowser
import time

# Configura uma pausa padrão entre cada comando do PyAutoGUI
# Isso ajuda a evitar que o bot "atropele" o carregamento do navegador
pyautogui.PAUSE = 0.5

def preencher_formulario():
    # 1. Abrir a página no navegador padrão
    url = "http://127.0.0.1:5500/index.html"
    webbrowser.open(url)
    
    # Aguarda o navegador abrir e carregar a página
    time.sleep(3) 

    # 2. Colocar o foco no primeiro campo (Nome)
    # Dependendo do navegador, um ou dois Tabs podem ser necessários.
    # Vamos assumir que o foco inicial começa fora do formulário.
    pyautogui.press('tab')

    # 3. Preencher os campos sequenciais
    # Nome
    pyautogui.write('Gilvan Daniel')
    pyautogui.press('tab')

    # Email
    pyautogui.write('gilvan.silva@email.com')
    pyautogui.press('tab')

    # CPF
    pyautogui.write('123.456.789-00')
    pyautogui.press('tab')

    # Telefone
    pyautogui.write('92991234567')
    pyautogui.press('tab')

    # Endereço
    pyautogui.write('Avenida Darcy Vargas, 1200')
    pyautogui.press('tab')

    # Cidade
    pyautogui.write('Manaus')
    pyautogui.press('tab')

    # 4. Seletores (Estado e Categoria)
    
    # Estado: Selecionar "AM"
    # Pressionamos a seta para baixo até chegar na opção desejada
    # (Supondo que AM seja a primeira ou que "A" leve até ela)
    pyautogui.write('am') 
    pyautogui.press('tab')

    # CEP
    pyautogui.write('69050-000')
    pyautogui.press('tab')

    # Categoria: Selecionar "Aluno"
    # Usando a lógica de setas conforme solicitado
    pyautogui.press('down') # Seleciona a primeira opção/próxima
    pyautogui.press('tab')

    # 5. Observações (O cursor deve parar aqui)
    pyautogui.write('Automacao concluida com sucesso pelo bot Python.')

if __name__ == "__main__":
    preencher_formulario()
```
